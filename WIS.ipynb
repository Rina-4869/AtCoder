````python
n = int(input())
w = list(map(int, input().split()))
#print(n, w)

    
def wis(n,w):
    A = [0]*(n+1)
    A[0]=0
    A[1]=w[0]
    
    for i in range(2, n+1):
        A[i] = max(A[i-1], A[i-2]+w[i-1])
    S = []
    i = n
    
    while i >= 1:
        if A[i] == A[i-1]:
            i -= 1
        else:
            S.append(i)
            i -= 2
    S.reverse()     
    return ([A[n],S])

ans=wis(n,w)
print("weight = "+ str(ans[0]))
print("vertices = "+ str(ans[1]))
````
