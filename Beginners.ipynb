````python
n = int ( input())
A=list (map(int, input().split()))

def count_div2(x: int) -> int:
    cnt = 0
    while x% 2 ==0:
        x//=2
        cnt+=1
    return cnt

ans=min(count_div2(a) for a in A)
print(ans)
        
````

同じ種類の硬貨どうしは区別できません。2 通りの硬貨の選び方は、ある種類の硬貨についてその硬貨を選ぶ枚数が異なるとき区別されます。

日本でよく使われる紙幣は、
10000 円札、

青橋くんが言うには、彼が祖父から受け取ったお年玉袋にはお札が 
N 枚入っていて、合計で 
Y 円だったそうですが、嘘かもしれません。このような状況がありうるか判定し、ありうる場合はお年玉袋の中身の候補を一つ見つけてください。なお、彼の祖父は十分裕福であり、お年玉袋は十分大きかったものとします。

制約
1≤N≤2000
1000≤Y≤2×10 
7
 
N は整数である。

````python
n, y = map(int, input().split())
y //= 1000  # 1000円単位にする

for i in range(n + 1):                 # 10000円札の枚数
    for j in range(n - i + 1):         # 5000円札の枚数
        k = n - i - j                  # 1000円札の枚数
        if 10*i + 5*j + 1*k == y:
            print(i, j, k)
            exit()

print(-1, -1, -1)
````

## endswith()

Pythonの文字列メソッド（str method）

**意味**：
「その文字列が、指定した文字列で終わっているか」を調べる関数。

---

## 基本形

```python
s.endswith("abc")
```

意味：

> 文字列 s の末尾が `"abc"` なら True、そうでなければ False

---

## 例

```python
s = "dreamer"

print(s.endswith("er"))     # True
print(s.endswith("dream"))  # False
```

---

## ABC049C での使い方

```python
if s.endswith("dream"):
    s = s[:-5]
```

意味：

> もし s が "dream" で終わっていたら、その部分を削る

---

## startswith() との対比（重要）

* **startswith()**：先頭か？

```python
s.startswith("dream")
```

* **endswith()**：末尾か？

```python
s.endswith("dream")
```

---

## 戻り値

* True（終わっている）
* False（終わっていない）

if文でこう使う：


````python
s= input().strip()

words= [ "dream", "dreamer", "erase", "eraser"]

while len(s)>0:
    matched =False
    for w in words:
        if len(s) >= len(w) and s[-len(w):]==w:
            s=s[:-len(w)]
            matched = True
            break
    if not matched:
        print("NO")
        exit()
        
print("YES")
````
