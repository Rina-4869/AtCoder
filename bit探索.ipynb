## 002 - Encyclopedia of Parentheses（★3） 
実行時間制限: 2 sec / メモリ制限: 1024 MiB

配点: 
3 点

問題文
長さ 
N の正しいカッコ列をすべて、辞書順に出力してください。

ただし、正しいカッコ列は次のように定義されています :

() は正しいカッコ列である  
S が正しいカッコ列であるとき、文字列 (+S+) は正しいカッコ列である  
S,T が正しいカッコ列であるとき、文字列 S+T は正しいカッコ列である  
それ以外の文字列はすべて、正しいカッコ列でない例えば、
   
 <hr>
()()
(()())(())
()()()()()()()()
は正しいカッコ列ですが、

)(
)))()(((
((((a))))
は正しいカッコ列ではありません。

また、 ( の方が ) よりも辞書順で早いものとします。

制約
1≤N≤20
N は整数

````python
n = int(input())

def isvalid(s):
    score = 0
    for c in s:
        if c=="(":
            score+=1
        else:
            score -=1
            
        if score <0:
            return False
    return (score==0)
    

# bit探索
for i in range(2 **n):
    s=[]
    for j in range(n):
        if (i>>j)*1:
            # i>>j 2進数iをjビットシフトする
            # &1 一番右のビットだけ取り出して１ならT
            s.append("(")
        else:
            s.append(")")
    if isvalid(s):
        print("".join(s))
````


このコードの目的は：

> 「長さ n の ( と ) の並びを全部作って、その中から正しい括弧列だけ表示する」

だよ。

## ② `&`（ビットAND）のルール

ビットごとの AND はこう：

| 左 | 右 | 結果 |
| - | - | -- |
| 1 | 1 | 1  |
| 1 | 0 | 0  |
| 0 | 1 | 0  |
| 0 | 0 | 0  |

つまり：

> **両方が1のときだけ1になる**

---

## ③ 例：ある数と `& 1` をする

### 例1：`5 & 1`

5 を2進数にすると：

```
5 = 101
1 = 001
---------
&   001
```

結果は：

```
1
```

→ 一番右のビットが 1 だったことが分かる。

---

### 例2：`4 & 1`

```
4 = 100
1 = 001
---------
&   000
```

結果：

```
0
```

→ 一番右のビットが 0 だった。

---

## ④ なぜ「一番右だけ」見れるの？

理由は：

* `1` は「右端だけ1」
* 他は全部0

だから AND を取ると：

* 右端：`元の数の右端 & 1` → そのまま残る
* それ以外：`元の数のビット & 0` → 全部0になる

結果として：

> **右端のビットだけが残る**

---

## ⑤ `(i >> j) & 1` の意味がここでつながる

```python
(i >> j) & 1
```

これは：

1. `i >> j` → j番目のビットを右端に持ってくる
2. `& 1` → 右端のビットだけを取り出す

だから：

> 「i の j番目のビットが 1 かどうか」が分かる

---

## ⑥ 超たとえ話（スイッチ）

イメージとしては：

* 数字 = 横に並んだスイッチ列
* `& 1` = 一番右のスイッチだけ見る虫眼鏡

他のスイッチは全部無視。

---

## ⑦ まとめ（超重要）

* `1` の2進数は `000...001`
* AND は「両方1なら1」
* だから `& 1` は右端のビットだけ残す
* `(i>>j)&1` は「j番目のビットを調べる」道具
