実行時間制限: 2 sec / メモリ制限: 256 MiB

配点 : 300 点

問題文 シカのAtCoDeerくんは二次元平面上で旅行をしようとしています。 AtCoDeerくんの旅行プランでは、時刻 0 に 点 (0,0) を出発し、 1 以上 N 以下の各 i に対し、時刻 t i​に 点 (x i​,y i​) を訪れる予定です。

AtCoDeerくんが時刻 t に 点 (x,y) にいる時、 時刻 t+1 には 点 (x+1,y), (x−1,y), (x,y+1), (x,y−1) のうちいずれかに存在することができます。 その場にとどまることは出来ないことに注意してください。 AtCoDeerくんの旅行プランが実行可能かどうか判定してください。

制約 1 ≤ N ≤ 10 5

0 ≤ x i​

≤ 10 5

0 ≤ y i​

≤ 10 5

1 ≤ t i​

≤ 10 5

t i​

< t i+1​( 1 ≤ i ≤ N−1) 入力は全て整数

````python
n= int(input())
# tn={time: (x,y)}
t, x, y =0,0,0
flg=True

for i in range (n):
    ti, xi, yi = map(int, input().split())
    
    dt = ti-t
    d=abs(xi-x)+abs(yi-y)
    
    if d>dt or (dt-d)% 2 !=0:
        flg=False
    
    t,x,y =ti,xi,yi
print("Yes" if flg else "No")
````

## ■ マンハッタン距離（Manhattan distance）とは

**定義：**

2点
[
(x_1, y_1), (x_2, y_2)
]
のマンハッタン距離は、

[
|x_1 - x_2| + |y_1 - y_2|
]

で計算する距離のこと。

---

## ■ なぜ「マンハッタン」？

ニューヨーク（マンハッタン）の街は、
縦と横の道が格子状（碁盤の目）になっている。

だから、

* 斜めに行けない
* 上下左右にしか移動できない

この移動距離の考え方がマンハッタン距離。

---

## ■ 直感的な意味

### 例1：

点 A = (1, 2)
点 B = (4, 6)

マンハッタン距離：
[
|4-1| + |6-2| = 3 + 4 = 7
]

これは、

* 右に3
* 上に4
  進む最短距離という意味。

---

## ■ ユークリッド距離との違い

### ユークリッド距離（普通の距離）：

[
\sqrt{(x_1-x_2)^2 + (y_1-y_2)^2}
]

### マンハッタン距離：

[
|x_1-x_2| + |y_1-y_2|
]

ユークリッド距離は「斜めOK」、
マンハッタン距離は「縦横のみ」。

---

## ■ この問題（ABC086C）でなぜマンハッタン距離？

問題文にこうある：

> 時刻 t に (x,y) にいるとき、
> t+1 には
> (x+1,y), (x-1,y), (x,y+1), (x,y-1)
> のいずれかに行ける

つまり：

* 1秒に1マス
* 上下左右のみ

だから、必要な最小移動回数は

[
|dx| + |dy|
]

になる。これがマンハッタン距離。

---

## ■ なぜ abs（絶対値）が必要？

例えば：

前：(2,5)
次：(2,1)

差は：

```python
yi - y = 1 - 5 = -4
```

距離は「4」なのに、符号付きだと -4 になる。
距離は常に正なので `abs()` が必要。

正しい式：

```python
abs(xi - x) + abs(yi - y)
```

---

## ■ よくあるバグ

君がやったやつ：

```python
d = abs(xi-x) + (yi-y)
```

これは：

* y方向がマイナスになると距離が壊れる
* テスト1個だけ落ちる典型ミス

---

## ■ まとめ（覚える形）

マンハッタン距離は：

```python
d = abs(x1 - x2) + abs(y1 - y2)
```

意味は：

> 上下左右にしか動けない世界での最短距離
